```python
# Fibonacci Series
def fibonacci(n):
    fib = [0, 1]  # Starting with the first two numbers of the series

    for i in range(2, n):
        next_fib = fib[i-1] + fib[i-2]  # Calculating the next Fibonacci number
        fib.append(next_fib)  # Adding the next Fibonacci number to the list
    return fib


fibonacci(10)
```

```python
# Nested Dictionaries and Lists
allGuests = {'Alice': {'apples': 5, 'pretzels': 12},
             'Bob': {'ham sandwiches': 3, 'apples': 2},
             'Carol': {'cups': 3, 'apple pies': 1}}

def totalBrought(guests, item):
  numBrought = 0

  for k, v in guests.items():
    numBrought = numBrought + v.get(item, 0)
  return numBrought

print('Number of things being brought:')
print(' - Apples         ' + str(totalBrought(allGuests, 'apples')))
print(' - Cups           ' + str(totalBrought(allGuests, 'cups')))
print(' - Cakes          ' + str(totalBrought(allGuests, 'cakes')))
print(' - Ham Sandwiches ' + str(totalBrought(allGuests, 'ham sandwiches')))
print(' - Apple Pies     ' + str(totalBrought(allGuests, 'apple pies')))
```

**8) Regex**:

```python
import re
```

```python
# recreando un patrón
phoneNumRegex = re.compile(r'\d\d\d-\d\d\d-\d\d\d\d')
mo = phoneNumRegex.search('My number is 415-555-4242.')

print('Phone number found: ' + mo.group())
```

```python
# compile + search
heroRegex = re.compile(r'Batman|Tina Fey')

mo1 = heroRegex.search('Batman and Tina Fey')
print(mo1.group())

mo2 = heroRegex.search('Tina Fey and Batman')
print(mo2.group())
```

```python
# findall v2
vowelRegex = re.compile(r'[aeiouAEIOU]')
print(vowelRegex.findall('RoboCop eats baby food. BABY FOOD.'))

consonantRegex = re.compile(r'[^aeiouAEIOU]')
print(consonantRegex.findall('RoboCop eats baby food. BABY FOOD.'))
```

**10) DateTime**:

```python
# datetime

import datetime
import time

halloween2024 = datetime.datetime(2024, 10, 31, 0, 0, 0)

while datetime.datetime.now() < halloween2024:
  print(datetime.datetime.now())
  time.sleep(1)
```

```python
# datetime v2
oct21st = datetime.datetime(2019, 10, 21, 16, 29, 0)

print(oct21st.strftime('%Y/%m/%d %H:%M:%S'))
print(oct21st.strftime('%I:%M %p'))
print(oct21st.strftime("%B of '%y"))
```

```python
# datetime v3
print(datetime.datetime.strptime('October 21, 2019', '%B %d, %Y'))
print(datetime.datetime.strptime('2019/10/21 16:29:00', '%Y/%m/%d %H:%M:%S'))
print(datetime.datetime.strptime("October of '19", "%B of '%y"))
print(datetime.datetime.strptime("November of '63", "%B of '%y"))
```
